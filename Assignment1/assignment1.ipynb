- 쿼카와 고양이

에이미라는 친구가 있습니다. 에이미는 그림 그리는 것을 좋아합니다.

특히 쿼카와 고양이 그림을 그리는 것을 좋아하는데요.

어느날 에이미는 자신이 그린 쿼카와 고양이 그림을 분류할 수 있는 딥러닝 모델을 만들고 싶어졌습니다.

그래서 intro-dl 세션에서 배운 것처럼 pre-trained ResNet50 모델을 사용해서 분류 모델을 만들었습니다.

아무래도 쿼카와 고양이는 정말 다르게 생긴만큼 에이미는 자신의 모델이 분류 작업을 정말 잘 해낼 것이라고 생각했습니다.

그런데 결과는 그렇게 좋지 않았습니다. 분류 모델의 성능이 약 80% 정도밖에 되지 않은 것입니다. 그래서 에이미는 이런 문제를 극복해보기로 했습니다.


이제 문제입니다. 에이미는 어떤 식으로 자신의 모델을 발전시켜야 할까요?

**1.   데이터 전처리**

가장 먼저, 데이터에 대해 의문을 가질 필요가 있습니다.

- 에이미의 쿼카와 고양이 그림이 서로 섞여 쿼카인데 고양이라고 잘못 라벨링이 되었다던가,

- 에이미가 쿼카라고 라벨링된 그림 귀퉁이에 고양이를 그렸을 수도 있습니다!

데이터를 잘 확인하는 것은 가장 중요한 작업 중 하나입니다.

**2.   모델 구조 변경**

에이미는 ResNet50 분류 모델을 사용했지만, 어쩌면 이 모델은 쿼카와 고양이 분류 같은 간단한 작업에 적합하지 않을 수 있습니다.

- 보다 간단한 CNN 모형을 설계해보면 더 좋은 결과를 낼 수도 있습니다.



이렇게 하나씩 중요한 사항들을 체크하면서 모델을 개선하면 됩니다.

하지만 이러한 작업이 놓치고 있는 것은 바로, "나"만의 아이디어를 담은 부분이 없다는 것입니다.

"나" 만의 새로운 아이디어를 추가해 모델을 어떻게 발전시킬 수 있을까요?
아래에 시도해볼만한 사항들을 적어보았습니다:

- ResNet50를 병렬로 2개 (이상) 쌓아서 각각의 output을 더하거나/곱하거나/output 각각에 learnable weight를 주고 합해 이미지를 분류할 수 있을까?

- GAN처럼 두 개의 ResNet50 구조를 서로 싸우게 만들어서 더 좋은 성능을 내도록 학습할 수 있을까?

**3.   새로운 학습 테크닉 만들기**

새로운 learning technique을 만들어서 딥러닝 모델을 개선할 수도 있습니다.
